```python
from ultralytics import YOLO

# Load the YOLOv8c model
model = YOLO("yolov9c.pt")  # Load a specific YOLOv9c model 

# Train the model with the --single-cls flag
results = model.train(data="/teamspace/studios/this_studio/YOLOv9/Scence-Text-Detection-4/data.yaml", epochs=100, imgsz=640, single_cls=True)

# Print the training results
print(results)
```

```python
model.save("yolov9c_100epoch.pt")
```

```python
from ultralytics import YOLO

# Load the trained model
model = YOLO("yolov9c_100epoch.pt")

# Validate the model on the validation dataset
results = model.val(data="/teamspace/studios/this_studio/YOLOv9_SivSreynoch/Scence-Text-Detection-4/data.yaml")

# Print results
print(results)
```
